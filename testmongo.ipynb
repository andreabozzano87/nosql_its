```python
pipeline = [
    {
        '$group': {
            '_id': '$Artist', 
            'mediaSpotifyPopularity': {
                '$avg': '$Spotify Popularity'
            }
        }
    }, {
        '$sort': {
            'mediaSpotifyPopularity': -1
        }
    }
]

list(collection.aggregate(pipeline=pipeline))
```

```python
aggregati = pd.DataFrame(list(collection.aggregate(pipeline=pipeline)))
aggregati
```
